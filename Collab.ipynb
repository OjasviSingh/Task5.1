```python

import pandas as pd
import matplotlib.pyplot as plt  # Import the matplotlib.pyplot module

Mock_Data = pd.read_csv('ExerciseData_combined_csv_user1.csv')



#compute stats
mean_distance=Mock_Data['distance'].mean()
median_calories=Mock_Data['calories'].median()
std_dev_distance=Mock_Data['calories'].std()
print (f"Mean distance:{mean_distance}")
print(f"Median calories:{median_calories}")
print(f"standard deviation of distance:{std_dev_distance}")



# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(Mock_Data['distance'], Mock_Data['calories'], alpha=0.5)
plt.xlabel('distance')
plt.ylabel('calories')
plt.title('Scatter Plot of Distance vs. calories')
plt.show()

```
